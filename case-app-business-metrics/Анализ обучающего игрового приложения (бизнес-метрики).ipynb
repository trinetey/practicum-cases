# Часть 1. Теоретические вопросы о бизнес-метриках

### 1. Дайте определение ARPPU. В результате изменений в продукте ARPPU снизился. Это хорошо или плохо?

Аббревиатура **ARPPU** расшифровывается как *Average Revenue Per Paying User* и значит "*средний доход на одного платящего пользователя*". Этим данная метрика принципиально отличается от **ARPU**, где рассчитывается средний доход от всех пользователей, платящих и нет.

**Ценность метрики ARPPU** прежде всего в том, что она позволяет понять, насколько лояльно платящая аудитория относится к ценам, установленным бизнесом. Также, она косвенно показывает удовлетворённость продуктом - за интересный продукт обычно готовы платить больше и регулярнее.

На первый взгляд может показаться, что то, что ARPPU снизился - это однозначно плохо. Но на самом деле, если мы знаем только этот показатель, то не можем судить хорошо или плохо его снижение. И вот почему.

Допустим, есть компания, в новое приложение которой пришло 1000 пользователей. 100 из них приобрели подписку стоимостью 50 рублей. Выручка составила 5000 рублей.

**ARPPU** считается по формуле *доход*/*количество платящих пользователей* = 5000/100 = **50 рублей** составила здесь данная метрика и именно столько с каждого платящего пользователя получила компания.

На следующий месяц бизнес решил повысить стоимость подписки до 100 рублей. Выручка осталась та же (5000), но количество пользователей, купивших продление подписки по новой цене, сократилось пропорционально повышению вдвое - до 50 пользователей.

**ARPPU по новым данным** составил: 5000/50 = **100 рублей**.

ARPPU вырос, но толку от этого мало, так как ушла половина платящих пользователей. Если бы их ушло больше ещё хотя бы на одного пользователя, то относительно предыдущего месяца этот период показал бы снижение дохода. Кроме того, в любом случае перед бизнесом встаёт вопрос поиска новых платящих пользователей взамен ушедшим.

В это же время, **снизившийся ARPPU**, напротив, **может показывать то, что у продукта прибавилось лояльных пользователей**. Вернемся к нашему примеру выше.

Компания решила не повысить цену подписки, а снизить на 5 рублей. Это позволило привлечь уже не 100, а 120 платящих пользователей, что принесло 120 Х 45 = 5400 рублей дохода.

В итоге ARPPU = 5400/120 = 45. Это меньше 50 - т.е. метрика снизилась относительно первоначального примера. Но видим рост дохода на +400, видим увеличение количества платящих пользователей, что в перспективе важнее для бизнеса.

Но если бы платящих клиентов пришло меньше? Не 120, а 110. Пользователей всё равно стало бы больше. ARPPU был бы те же 45 рублей против 50. Но и прибыль бы мы получили меньшую - 4950, что, опять же, не очень хорошо.

Эти примеры наглядно объясняют почему **нельзя рассматривать ARPPU в отрыве от остальных показателей**. Для определения того, хорошо или плохо то, что данная метрика снизилась, нам как минимум нужны данные по выручке, ARPU и Paying Share (доля платящих от всей аудитории). Тогда будет понятно негативно ли пользователи отнеслись к изменениям в продукте (в таком случае на понижение пойдут и другие метрики, не только ARPPU) или просто в продукт были добавлены новые более бюджетные опции, с которых получается более низкий ARPPU, но при этом клиентов, платежей и денег у приложения больше.

### 1.2 Пишем функцию для составления профилей пользователей

Есть в наших данных один важный момент - даты первой сессии некоторых пользователей старше даты регистрации. К примеру, вот у этого пользователя регистрация по записям произошла аж на сутки позже, чем первая сессия (16 мая в 02:05:38 первая сессия, а регистрация только 17-го числа в 03:22:17):

```python
users.query('user_id == 622739')
```

```python
sessions.query('user_id == 622739').head(5)
```

Почему так получилось? Вероятно, имел место технический сбой, из-за чего данные некорректно записались в датафреймы.

Таких аномальных пользователей мало (около 2% от всех ID), но если их не исключить, то они не дадут нам точно рассчитать удержание.

Исключение проведём на этапе создания профилей пользователей:

```python
def get_profiles(users, payments):
    # сортируем сессии по ID пользователя и дате посещения, группируем по ID и агреригуем reg_time
    # столбец с временем первого посещения назовём registration
    profiles = (
        users.sort_values(by=['user_id', 'reg_time'])
        .groupby('user_id')
        .agg({'reg_time': 'first'})
        .rename(columns={'reg_time': 'registration'})
        .reset_index()  # возвращаем user_id из индекса
    )

    # создадим табличку с датами первых сессий
    first_visits = (
         sessions.sort_values(by=['user_id', 'open_time'])
        .groupby('user_id')
        .agg({'open_time': 'first'})
        .rename(columns={'open_time': 'first_visit'})
        .reset_index()
    )

    # объединим табличку с датами первых сессий с профилями
    users_pivot = (
        profiles.merge(first_visits[['user_id', 'first_visit']], on='user_id', how='left')
    )

    # выделим отдельно дату первой сессии и дату регистрации
    users_pivot['first_session_dt'] = users_pivot['first_visit'].astype('datetime64[D]')

    users_pivot['dt'] = users_pivot['registration'].astype('datetime64[D]')

    # фильтруем профили так, чтобы остались только те, чья регистрация была раньше или равна по времени первой сессии, также
    # отфильтровываем те строки, где дата регистрации меньше, чем дата первой сессии
    profiles = users_pivot.query('dt >= first_session_dt and registration <= first_visit')

    # При помощи isin определяем, есть ли id пользователя среди покупателей
    profiles['payer'] = profiles['user_id'].isin(payments['user_id'].unique())

    # Добавляем в профили данные по уровням, на котором остановился пользователь:
    profiles = (
        profiles.merge(last_level[['user_id', 'level_group']], on='user_id', how='left')
    )

    return profiles

profiles = get_profiles(users, payments)

# Проверяем отработку функции:
profiles
```

Профили отфильтрованы. Теперь в них есть только корректные даты регистрации и первого посещения, а также добавлен новый столбец с индикатором актуального уровня.

### 1.4 Пишем функции для расчёта и визуализации удержания (retention)

```python
def get_retention(
    profiles,
    sessions,
    horizon_days,
    observation_date,
    dimensions=[],
    ignore_horizon=False
):

    # добавляем столбец payer в передаваемый dimensions список
    dimensions = ['payer'] + dimensions

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )

    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # собираем «сырые» данные для расчёта удержания
    result_raw = result_raw.merge(
        sessions[['user_id', 'open_time']], on='user_id', how='left'
    )

    result_raw['lifetime'] = (
        result_raw['open_time'] - result_raw['registration']
    ).dt.days

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        # расчет удержания
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        # вычисляем размеры когорт
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        # добавляем размеры когорт в таблицу
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        result = result.div(result['cohort_size'], axis=0)
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        # восстанавливаем размеры когорт
        result['cohort_size'] = cohort_sizes
        return result

    # получаем таблицу удержания
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # получаем таблицу динамики удержания
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )

    # возвращаем обе таблицы и сырые данные
    return result_raw, result_grouped, result_in_time

retention_raw, retention, retention_history = get_retention(
    profiles, sessions, 10, datetime(2023, 5, 31).date()
)

# проверяем работу функции:
retention
```

Функция работает. Удержание нулевого дня рассчиталось верно - оно равно единице, чего не было бы, если бы мы не отфильтровали данные с неккоректными датами.

Напишем, также, функцию для визуализации удержания:

```python
def plot_retention(retention, retention_history, horizon):

    # задаём размер сетки для графиков
    plt.figure(figsize=(15, 10))

    # исключаем размеры когорт и удержания первого дня
    retention = retention.drop(columns=['cohort_size', 0])

    # если в индексах таблицы удержания только payer,
    # добавляем второй признак — cohort
    if retention.index.nlevels == 1:
        retention['cohort'] = 'All users'
        retention = retention.reset_index().set_index(['cohort', 'payer'])

    # в таблице графиков — два столбца и две строки, четыре ячейки
    # в первой строим кривые удержания платящих пользователей
    ax1 = plt.subplot(2, 2, 1)
    retention.query('payer == True').droplevel('payer').T.plot(
        grid=True, ax=ax1
    )
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание платящих пользователей')

    # во второй ячейке строим кривые удержания неплатящих
    # вертикальная ось — от графика из первой ячейки
    ax2 = plt.subplot(2, 2, 2, sharey = ax1)
    retention.query('payer == False').droplevel('payer').T.plot(
        grid=True, ax=ax2
    )
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание неплатящих пользователей')

    # в третьей ячейке — динамика удержания платящих
    ax3 = plt.subplot(2, 2, 3)
    # получаем названия столбцов для сводной таблицы
    columns = [
        name
        for name in retention_history.index.names
        if name not in ['dt', 'payer']
    ]
    # фильтруем данные, строим сводную таблицу и график по ней
    filtered_data = retention_history.query('payer == True').pivot_table(
        index='dt', columns=columns, values=horizon-1, aggfunc='mean'
    )
    filtered_data.plot(grid=True, ax=ax3)
    plt.xlabel('Дата привлечения')
    plt.title(
        'Динамика удержания платящих пользователей на {}-й день'.format(
            horizon
        )
    )

    # в четвёртой ячейке — динамика удержания неплатящих
    ax4 = plt.subplot(2, 2, 4, sharey = ax3)
    # фильтруем данные, строим сводную таблицу и график по ней
    retention_history.query('payer == False').pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    ).plot(grid=True, ax=ax4)
    plt.xlabel('Дата привлечения')
    plt.title(
        'Динамика удержания неплатящих пользователей на {}-й день'.format(
            horizon
        )
    )

    # исключаем наложение графиков друг на друга
    plt.tight_layout()

    plt.show()
```

### 1.5 Функции для расчета и визуализации LTV

Вообще LTV рекомендуется считать на базе дохода. Но в рамках этого проекта мы посмотрим также и эту метрику, предположив, что у приложения нет или почти нет расходов на продвижение/содержание и, таким образом, представим, что выручка - это и есть доход.

Для того, чтобы LTV дало нам полезную информацию, нужно прописать в нем деление на платящих и не платящих пользователей. Без этого деления неплатящие, коих очень много, "смажут" итоговую картину.

Подготовим функцию для расчёта LTV.

```python
# Считаем LTV:
def get_ltv(
    profiles,
    payments,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):
    # пропишем деление на платящих/не платящих:
    dimensions = ['payer'] + dimensions

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')
    # добавляем данные о покупках в профили
    result_raw = result_raw.merge(
        payments[['user_id', 'pay_time', 'amt']], on='user_id', how='left'
    )
    # рассчитываем лайфтайм пользователя для каждой покупки
    result_raw['lifetime'] = (
        result_raw['pay_time'] - result_raw['registration']
    ).dt.days

    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort']

    # функция группировки по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        # строим «треугольную» таблицу выручки
        result = df.pivot_table(
            index=dims, columns='lifetime', values='amt', aggfunc='sum'
        )
        # находим сумму выручки с накоплением
        result = result.fillna(0).cumsum(axis=1)
        # вычисляем размеры когорт
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        # объединяем размеры когорт и таблицу выручки
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # считаем LTV: делим каждую «ячейку» в строке на размер когорты
        result = result.div(result['cohort_size'], axis=0)
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        # восстанавливаем размеры когорт
        result['cohort_size'] = cohort_sizes

        # возвращаем таблицу LTV
        return result

    # получаем таблицу LTV
    result_grouped = group_by_dimensions(
        result_raw, dimensions, horizon_days
    )

    # для таблиц динамики убираем 'cohort' из dimensions
    if 'cohort' in dimensions:
        dimensions = []

    # получаем таблицы динамики LTV
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )

    return (
        result_raw,  # сырые данные
        result_grouped,  # таблица LTV
        result_in_time,  # таблица динамики LTV
    )

# Вызываем функцию для расчета LTV и сырых данных
ltv_raw, ltv_grouped, ltv_history = get_ltv(
    profiles, payments, datetime(2023, 5, 30).date(), 10
)

# Проверим отработку функции
display(ltv_grouped)
```

А теперь напишем функцию визуализации:

```python
def plot_ltv(ltv, ltv_history, horizon):

    # задаём сетку отрисовки графиков
    plt.figure(figsize=(20, 8))

    # из таблицы ltv исключаем размеры когорт
    ltv = ltv.drop(columns=['cohort_size'])
    # в таблице динамики ltv оставляем только нужный лайфтайм
    ltv_history = ltv_history.drop(columns=['cohort_size'])[[horizon - 1]]

    # первый график — кривые ltv
    ax1 = plt.subplot(2, 3, 1)
    ltv.T.plot(grid=True, ax=ax1)
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('LTV')

    # второй график — динамика ltv
    ax2 = plt.subplot(2, 3, 2, sharey=ax1)
    # столбцами сводной таблицы станут все столбцы индекса, кроме даты
    columns = [name for name in ltv_history.index.names if name not in ['dt']]
    filtered_data = ltv_history.pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filtered_data.plot(grid=True, ax=ax2)
    plt.xlabel('Дата привлечения')
    plt.title('Динамика LTV пользователей на {}-й день'.format(horizon))

    plt.tight_layout()
    plt.show()

# считаем LTV
ltv_raw, ltv_grouped, ltv_history = get_ltv(
    profiles, payments, datetime(2023, 5, 31).date(), 10
)
```

### 1.6 Функции для расчета и визуализации конверсии

```python
# Получаем пользовательские профили и данные о покупках
# передаём их в качестве аргументов profiles и orders
def get_conversion(
    profiles,
    payments,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # определяем дату и время первой покупки для каждого пользователя
    first_purchases = (
        payments.sort_values(by=['user_id', 'day_pay'])
        .groupby('user_id')
        .agg({'day_pay': 'first'})
        .reset_index()
    )

    # добавляем данные о покупках в профили
    result_raw = result_raw.merge(
        first_purchases[['user_id', 'day_pay']], on='user_id', how='left'
    )

    # рассчитываем лайфтайм для каждой покупки
    result_raw['lifetime'] = (
        result_raw['day_pay'] - result_raw['registration']
    ).dt.days

    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort']

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        # строим «треугольную» таблицу конверсии
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        # считаем сумму с накоплением для каждой строки
        result = result.fillna(0).cumsum(axis = 1)
        # вычисляем размеры когорт
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        # добавляем размеры когорт в таблицу конверсии
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # делим каждую «ячейку» в строке на размер когорты
        # и получаем conversion rate
        result = result.div(result['cohort_size'], axis=0)
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        # восстанавливаем размеры когорт
        result['cohort_size'] = cohort_sizes
        return result

    # получаем таблицу конверсии
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # для таблицы динамики конверсии убираем 'cohort' из dimensions
    if 'cohort' in dimensions:
        dimensions = []

    # получаем таблицу динамики конверсии
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )

    # возвращаем обе таблицы и сырые данные
    return result_raw, result_grouped, result_in_time

# считаем конверсию без параметра dimensions
conversion_raw, conversion, conversion_history = get_conversion(
    profiles, payments, datetime(2023, 5, 31).date(), 10
)
# Проверяем отработку функции:
conversion
```

### 2.3 Изучим, за какое время пользователи переходят с уровня на уровень

Поскольку пользователь может прийти, поиграть, уйти, а потом зайти в приложение через день-два, мы не можем просто вычесть разницу между датами достижения уровней - получится результат, словно бы пользователь 1-2 дня пытался пройти уровень. Поэтому для вычисления нам понадобится объединить столбцы с уровнями и сессиями и провести некоторые манипуляции, чтобы время игры на уровне вычислялось именно во время сессии игрока, игнорируя время, когда он явно не в приложении.

Для этого сделаем следующее:

```python
# объединяем сессии с уровнями:
sessions_level = levels.merge(sessions[['user_id', 'open_time', 'close_time']], on=['user_id'], how='left')
sessions_level
```

```python
# для избавления от дублей времени достижения уровня фильтруем датафрейм так, чтобы время прохождения уровня было больше или
# равно открытию сессии и при этом меньше, чем завершение сессии. Тогда время прохождения уровня окажется как раз напротив
# значений той сессии, в которой этот уровень был достигнут:
sessions_level = (
    sessions_level.loc[(sessions_level['time'] >= sessions_level['open_time'])
                       & (sessions_level['time'] <= sessions_level['close_time'])]
)
sessions_level
```

```python
# создаем столбец time_level, в котором будет отражено время прохождения уровня. Для этого группируем пользователей по id,
# старту сессии и её окончанию. Вычисляем разницу между временем прохождения при помощи diff. Так у нас заполнятся все значения
# в столбце, кроме тех, где уровень был первым или единственным за сессию - тут будут пропуски:
sessions_level['time_level']= sessions_level.groupby(['user_id', 'open_time', 'close_time'])['time'].diff()
sessions_level
```

Пропуски обработаем так:
- Создадим переменные:
    1. **y** - разница времени между завершением уровня и достижением уровня: игроки почти всегда достигают какого-то уровня, после чего ещё немного продолжают играть до завершения сессии. Это время тоже надо приплюсовать к тому, которое игрок потратит для достижения следующего уровня в следующей сессии. Но мы не учитываем тут 2 уровень - эти места остаются пустыми, чтобы на них не перешла разница от предыдущих игроков и время для каждого игрока таким образом считалось бы отдельно;
    2. **х** - разница между стартом сессии и временем достижения уровня.
    
Чтобы у нас верно обработались все пропуски, сначала заполним пустые места везде, кроме второго уровня комбинацией **x+y**. После чего обработаем и 2 уровень просто **х**.

```python
y = (sessions_level.query('level != 2')['close_time'] - sessions_level.query('level != 2')['time']).shift(1)
x = sessions_level['time'] - sessions_level['open_time']

# обрабатываем пропуски:
sessions_level['time_level'] = sessions_level['time_level'].fillna(x+y)
sessions_level['time_level'] = sessions_level['time_level'].fillna(x)

sessions_level
```

Время разницы между уровнями считается так:
- на 2 уровне: действует разница между `open_time` и `time`;
- на последующих уровнях, достигнутых в ЭТОЙ ЖЕ сессии - разница между временем уровней;
- на следующем (не втором) уровне в следующей сессии: (`time` - `open_time`) этой сессии + (`close_time` - `time`) предыдущей сессии.

Так выше, к примеру, мы видим, что у пользователя с `user_id` 6906 был достигнут 7 уровень примерно в 22:06, после чего он ещё 10 минут играл, прежде чем ушел. Эти 10 минут переносятся на следующую сессию, когда он достигает 8 уровня: там разница между открытием сессии и достижением уровня - около 1 часа 3 минут. Прибавляем к этому времени ранее накопленные 10 минут - их же пользователь тоже провел на 7 уровне, добираясь до 8-го, и получаем, что всего ~01:13 провел пользователь на уровне 7, пока добирался до 8 уровня.

```python
# Посчитаем медиану прохождения для каждого уровня:
level_duration = sessions_level.groupby('level')['time_level'].median().dt.round('1S')

# Делаем понятную сводную табличку:
level_duration = (
    level_duration.reset_index()
    .sort_values(by='time_level', ascending=False)
    .rename(columns={'level': 'Уровень', 'time_level': 'Медианное время игроков на уровне'})
    .style.hide(axis='index')
)
level_duration
```

По таблице видим довольно ожидаемую картину - уровни повыше пользователи проходят дольше, чем начальные. До 10-го уровня включительно пользователи тратят на прохождение 1-го уровня около 20 минут. Далее время прохождения имеет тенденцию к росту.

### 3.3 Другие монетизационные метрики

Изучим:

- **Paying users** - пользователи, заплатившие хотя бы раз;
- **% Paying users** - отношение платящих пользователей к общем количеству пользователей за период;
- **Revenue** - выручка;
- **Дневное количество покупок**;
- **AOV** (Average Order Value) - средний чек с каждого заказа;
- **RPR** (Repeat Purchase Rate) - процент покупателей, которые совершают повторную покупку;
- **ARPDAU** (Average Revenue per Daily Active User) - средний доход с одного уникального "дневного" пользователя;
- **ARPPU** (Average Revenue Per Paying User) - средний доход на одного платящего пользователя;

А также рассчитаем дополнительные пользовательские метрики, нужные нам для расчетов и анализа:
- **DAU** (Daily Active Users) - количество уникальных пользователей за сутки;
- **Sticky Factor** ('липкость' приложения) - показатель, который отражает регулярноcть взаимодействия пользователя с продуктом в течении выбранного периода времени.

#### 3.3.1 Для начала напишем функцию, чтобы сагрегировать данные из датафрейма **Payments**:

```python
def monetization(x):
    result = {
        'Paying users': x['user_id'].nunique(),
        'Revenue': x['amt'].sum(),
        'AOV': x['amt'].sum() / x['amt'].count(),
        'RPR': (x['user_id'].count() - x['user_id'].nunique()) / x['user_id'].count() * 100,
        'ARPPU': x['amt'].sum() / x['user_id'].nunique(),
        'Количество покупок': x['amt'].count()
    }
    return pd.Series(result, index=['Paying users', 'Revenue', 'AOV', 'RPR', 'ARPPU', 'Количество покупок']).round(2)

monetization_metrics = payments.groupby('day_pay').apply(monetization)
```

В получившуюся табличку добавим DAU, процент платящих пользователей и ARPDAU:

```python
monetization_metrics['DAU'] = sessions.groupby('dt_sessions')['user_id'].nunique()
monetization_metrics['% Paying users'] = ((monetization_metrics['Paying users'] / monetization_metrics['DAU']) * 100).round(2)
monetization_metrics['ARPDAU'] = (payments.groupby('day_pay')['amt'].sum() / monetization_metrics['DAU']).round(4)

# Расположим стоблцы более логично:
monetization_metrics = monetization_metrics[list(('DAU', 'Paying users', '% Paying users', 'Revenue', 'Количество покупок',
                                                  'AOV', 'RPR', 'ARPDAU', 'ARPPU'))]
monetization_metrics
```

### 3.4 "Липкий" фактор

Зная DAU и рассчитав WAU (количество уникальных пользователей в неделю) мы, также, можем рассчитать ещё одну метрику - "липкий фактор". Она покажет нам как регулярно пользуются сервисом наши пользователи в течение недели.

```python
# Вычисляем среднюю дневную посещаемость уникальными пользователями
dau_total = sessions.groupby('dt_sessions')['user_id'].nunique().mean()

# Считаем среднюю недельную посещаемость уникальными пользователями
wau_total = sessions.groupby('week')['user_id'].nunique().mean()

# По формуле считаем "липкий фактор"
sticky_wau = (dau_total/wau_total) * 100

# Выводим показатель на экран:
round((sticky_wau), 2)
```

К этой метрике, учитывая, что у нас в датасете не такой большой объем данных по временному интервалу, нужно относится осторожно. Но навскидку можно отметить, что показатель "липкости2 неплохой. Аудитория приложения явно лояльна и заинтересована в нём. К сожалению, при этом она не слишком спешит платить, как мы выяснили выше. ### 3.5 Посчитаем, где ученики чаще всего впервые платят

#### 3.5.1 Выясним уровень, где чаще всего совершают первый платеж

```python
# объединяем таблицы с платежами и уровнями
pay_level = payments.merge(levels[['user_id', 'level', 'time']], on=['user_id'], how='left')

# ставим фильтр так, чтобы время достижения уровня было меньше даты платежа - так мы в дальнейшем сможем получить строки именно
# с теми уровнями, на которых находился пользователь во время совершения платежа, а не уровень выше:
pay_level = pay_level.loc[(pay_level['pay_time'] >= pay_level['time'])]

# считаем уровень, где был самый первый платеж:
first_pay = (
    pay_level.sort_values(['user_id', 'pay_time', 'level'])
    .groupby(['user_id', 'pay_time'])
    .agg({'level': 'max'}) # получаем актуальные уровни на момент платежей
    .groupby('user_id')['level'].min() # получаем минимальный из уровней, где были платежи
    .reset_index()
    .rename(columns={'level': 'Уровень'})
    .groupby('Уровень')['Уровень'].count() #считаем платежи по уровням
)

# Строим график:
plt.figure(figsize=(15,6))
g = sns.barplot(x=first_pay.index, y=first_pay.values, color='#99D9EA', edgecolor='#BB8FCE')
g.set_xlabel('Уровень', fontsize=15)
g.set_ylabel('Количество платежей', fontsize=15)
plt.title('На каких уровнях пользователи совершают первые платежи?', fontsize=18)
plt.show()
```

Больше всего пользователей делает первые платежи на 8 уровне. Также, неплохо толкают на первую покупку 5, 9, 4 и 7 уровни.

Кроме того, можно отметить, что пользователи решаются на первую покупку в первые 12 уровней игры.

### 3.6 На каком уровне ученики совершают больше всего платежей?

```python
# находим те уровни, на которых находились пользователи в момент, когда решили заплатить:
most_pay = (
    pay_level.sort_values(['user_id', 'pay_time', 'level'])
    .groupby(['user_id', 'pay_time'])
    .agg({'level': 'max'})
    .rename(columns={'max': 'Уровень'})
    .reset_index()
)

# считаем, на каком уровне сколько платежей:
most_paid_level = most_pay.groupby('level')['level'].count()

# Строим график:
plt.figure(figsize=(15,6))
g = sns.barplot(x=most_paid_level.index, y=most_paid_level.values, color='#99D9EA', edgecolor='#BB8FCE')
g.set_xlabel('Уровень', fontsize=15)
g.set_ylabel('Количество платежей', fontsize=15)
plt.title('На каких уровнях чаще всего платят пользователи?', fontsize=18)
plt.show()
```

Видим, что чаще всего платят на 8, 9 и 10-ом уровнях. Возможно, их сложно проходить без дополнительного вложения. А из начальных уровней самые оплачиваемые 4 и 5.
