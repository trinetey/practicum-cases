### Задание 3.3 Лучший день недели для рассылок

Чтобы выяснить лучший день для рассылок, сгруппируем таблицу по дням недели, в которые делались рассылки и рассчитаем медианные значения метрик Open Rate, CTOR и UR в эти дни. Берем медиану, так как она в отличие от среднего, нечувствительна к выбросам.

```python
df.groupby('День недели.1')[['% open_rate', '% ctor', '% ur']].median()
```

Видим, что по совокупности метрик лучшие показатели демонстрируют суббота и среда.

При этом у субботы чуть выше метрики Open Rate, CTOR и одновременно с этим UR ниже, чем в среду. Кажется победитель очевиден, однако посмотрим сколько вообще рассылок отправлялось в разные дни:

```python
df.groupby('День недели.1')['День недели.1'].count()
```

Видим, что рассылок в субботу за весь год было всего 5 штук, а это значит, что куда больше доверия вызывают данные по среде, где их было 40.
